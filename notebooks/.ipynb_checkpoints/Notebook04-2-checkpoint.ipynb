# Hybrid assembly of <em>E. coli</em> genome

Let's continue where we left off. Let's just check the device is mounted beforehand, and that we are in EColi directory.

```bash
cd EColi
```

Okay, good. Get the raw data. wget is a bit slow, so I used aria2, which is a supposedly multithreaded faster downloader. I have a list of files to download in list_download.txt

```bash
cat list_download.txt
```

```bash
aria2c -i list_download.txt -x 10 -s 10
```

We have not subsampled the long reads yet. Based on my simple calculations, the first 5181 reads suffice (definitely not a scientific way of subsampling).

```bash
head -10362 ecoli_allreads.fasta > cov30x.fasta
```

Long read assembly first.

```bash
unicycler -l cov30x.fasta \
 -o nanopore_only \
 -t 64
```

5 minutes! That is amazing! Continuing with hybrid assembly.

```bash
unicycler -1 s_6_1.fastq.gz \
 -2 s_6_2.fastq.gz \
 -l cov30x.fasta \
 -o hybrid_assembly \
 -t 64
```
