# RStudio Server

Personally, it is quite complicated to set up our RStudio server. First, add CRAN repository entry into our repository list. We are using Ubuntu 18 - the Bionic Beaver, so we getting a repo entry for Bionic. Check which entry you require [here](https://cran.rstudio.com/bin/linux/ubuntu/README.html).

```bash
echo "deb https://cloud.r-project.org/bin/linux/ubuntu bionic-cran35/" \
 | sudo tee -a /etc/apt/sources.list
```

Update apt repo list.

Then, install R.

```bash
sudo apt install r-base -y
```

Next, install gdebi-core, download RStudio Server, then install RStudio Server

```bash
sudo apt-get install gdebi-core -y
```

```bash
wget https://download2.rstudio.org/server/bionic/amd64/rstudio-server-1.2.1335-amd64.deb
sudo gdebi rstudio-server-1.2.1335-amd64.deb
```

Now, go to your favourite web browser and type:

<pre>http://[ip address of AWS instance]:8787</pre>

![](../images/Notebook05-2/image01.png)

Key in the necessary information - unfortunately we don't have a user in our instance. Go back to our instance, create an account and set up it's password

I am pretty sure this was all the things that I typed. I tried to access RStudio server again and thank God it worked.
